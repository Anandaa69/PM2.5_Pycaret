```python
from pycaret.time_series import *
import pandas as pd
```

```python
model = load_model('temp_eng')
```

```python
predict = predict_model(model)
```

```python
predict
```

```python
model2 = load_model('humi_eng')
```

```python
predict2 = predict_model(model2)
```
